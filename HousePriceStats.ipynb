```python
postcode_lookup_dict = mergey[['flat_postcode', 'PAR22CD']].drop_duplicates().set_index('flat_postcode').to_dict()['PAR22CD']
last_year_onwards = dict((key, []) for key in oa_to_parish_lookup.PAR22CD.drop_duplicates())
last_three_years = dict((key, []) for key in oa_to_parish_lookup.PAR22CD.drop_duplicates())
last_five_years = dict((key, []) for key in oa_to_parish_lookup.PAR22CD.drop_duplicates())

ever_sold = dict((key, {
    'num_sold': 0,
    'total_sale_price': 0
}) for key in oa_to_parish_lookup.PAR22CD.drop_duplicates())

missing_list = []
```

```python
def string_nan(x):
    return x != x
```

```python
import math
chunksize=50000
sale_count = 0
missing_count = 0
columns = ['id', 'price', 'date', 'postcode', 'property_type', 'old_new', 'duration', 'paon', 'saon', 'street', 'locality', 'town', 'district', 'county', 'ppd_category', 'record_status']
i = 0
for chunk in pd.read_csv('data/pp-complete.csv', chunksize=chunksize, header=0, names=columns):
        df = chunk
        df['date'] = pd.to_datetime(df.date).apply(lambda x: x.date())
        for ind, row in df.iterrows():
            postcode = row['postcode']
            parish = None
            if not string_nan(postcode):
                parish = postcode_lookup_dict.get(re.sub('[\s+]', '', postcode), None)
            if not parish:
                missing_count += 1
                missing_list.append(postcode)
                continue
            else:
                sale_count += 1
            price = row['price']
            if row['date'] > datetime.date(2022,1,1):
                last_year_onwards[parish].append(price)
            if row['date'] > datetime.date(2019,1,1):                
                last_three_years[parish].append(price)
            if row['date'] > datetime.date(2017,1,1):                
                last_five_years[parish].append(price)         
            ever_sold[parish]['num_sold'] += 1
            ever_sold[parish]['total_sale_price'] += price
        if i % 50 == 0:
            print("Done chunk {0}".format(i))
        i += 1
```

```python
better_stats = []
for frame in [last_year_onwards, last_three_years, last_five_years]:
    dicty = {}
    for parish, sales in frame.items():
        median = np.median(sales)
        number = len(sales)
        dicty[parish] = {'median_price': median, 'total_sales': number}
    better_stats.append(dicty)
    
```
